```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Check if GPU is available; use it if available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Define the neural network architecture
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.hidden_layer1 = nn.Linear(1, 128)
        self.hidden_layer2 = nn.Linear(128, 128)
        self.hidden_layer3 = nn.Linear(128, 128)
        self.output_layer = nn.Linear(128, 3)
        self.tanh = nn.Tanh()

    def forward(self, eta):
        x = self.tanh(self.hidden_layer1(eta))
        x = self.tanh(self.hidden_layer2(x))
        x = self.tanh(self.hidden_layer3(x))
        return self.output_layer(x)

# Define the ODE system
def ode_system(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max):
    eta.requires_grad = True
    y = model(eta)
    f, g, h = y[:, 0:1], y[:, 1:2], y[:, 2:3]

    # First derivatives
    f_prime = torch.autograd.grad(f, eta, grad_outputs=torch.ones_like(f), create_graph=True)[0]
    g_prime = torch.autograd.grad(g, eta, grad_outputs=torch.ones_like(g), create_graph=True)[0]
    h_prime = torch.autograd.grad(h, eta, grad_outputs=torch.ones_like(h), create_graph=True)[0]

    # Second derivatives
    f_double_prime = torch.autograd.grad(f_prime, eta, grad_outputs=torch.ones_like(f_prime), create_graph=True)[0]
    g_double_prime = torch.autograd.grad(g_prime, eta, grad_outputs=torch.ones_like(g_prime), create_graph=True)[0]
    h_double_prime = torch.autograd.grad(h_prime, eta, grad_outputs=torch.ones_like(h_prime), create_graph=True)[0]

    # Third derivatives
    f_triple_prime = torch.autograd.grad(f_double_prime, eta, grad_outputs=torch.ones_like(f_double_prime), create_graph=True)[0]

    ode1 = f_triple_prime + (n + 1) / 2 * f_prime * f_double_prime - n * f_prime ** 2 - M * f_prime - A * (f_prime + eta / 2 * f_double_prime)
    ode2 = g_double_prime + 0.5 * Pr * (n + 1) * f * g_prime + Nb * h_prime * g_prime + Nt * g_prime ** 2 - A / 2 * eta * g_prime
    ode3 = h_double_prime + (n + 1) / 2 * Ln * f * h_prime + Nt / Nb * g_double_prime - A * Ln / 2 * eta * h_prime
    return ode1, ode2, ode3

# Define the loss function
def loss_function(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max):
    eta.requires_grad = True
    y_pred = model(eta)
    f_pred, g_pred, h_pred = y_pred[:, :1], y_pred[:, 1:2], y_pred[:, 2:3]
    ode1, ode2, ode3 = ode_system(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max)

    # Boundary conditions at eta = 0 and eta = 10
    bc_eta_0 = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)
    bc_eta_10 = torch.tensor([[10.0]], dtype=torch.float32, requires_grad=True)
    bc_pred_0 = model(bc_eta_0)
    bc_pred_10 = model(bc_eta_10)

    f_bc0, g_bc0, h_bc0 = bc_pred_0[:, 0], bc_pred_0[:, 1], bc_pred_0[:, 2]
    f_prime_bc0 = torch.autograd.grad(f_bc0, bc_eta_0, grad_outputs=torch.ones_like(f_bc0), create_graph=True)[0]
    f_double_prime_bc0 = torch.autograd.grad(f_prime_bc0, bc_eta_0, grad_outputs=torch.ones_like(f_prime_bc0), create_graph=True)[0]
    f_prime_bc10 = torch.autograd.grad(bc_pred_10[:, 0], bc_eta_10, grad_outputs=torch.ones_like(bc_pred_10[:, 0]), create_graph=True)[0]
    g_prime_bc0 = torch.autograd.grad(g_bc0, bc_eta_0, grad_outputs=torch.ones_like(g_bc0), create_graph=True)[0]

    loss_bc1 = torch.mean((f_bc0 - 0) ** 2)
    loss_bc2 = torch.mean((f_prime_bc0 - 1 - gamma * f_double_prime_bc0) ** 2)
    loss_bc3 = torch.mean((g_bc0 - 1) ** 2)
    loss_bc4 = torch.mean((h_bc0 - 1) ** 2)
    loss_bc5 = torch.mean((f_prime_bc10 - 0) ** 2)
    loss_bc6 = torch.mean((bc_pred_10[:, 1] - 0) ** 2)
    loss_bc7 = torch.mean((bc_pred_10[:, 2] - 0) ** 2)

    # ODE loss
    loss_ode = torch.mean(ode1 ** 2) + torch.mean(ode2 ** 2) + torch.mean(ode3 ** 2)

    return loss_ode + loss_bc1 + loss_bc2 + loss_bc3 + loss_bc4 + loss_bc5 + loss_bc6 + loss_bc7

# Training data for eta
eta_train = torch.tensor(np.linspace(0, 10, 100).reshape(-1, 1), dtype=torch.float32, requires_grad=True)
A, M, n, Pr, Nb, Nt, Ln, gamma, e_max = 1, 0.5, 2, 1, 0.2, 0.2, 2, 0.1, 12

# Plotting function
def train_and_plot(parameter_values, parameter_name, model_params):
    colours = ['red', 'green', 'blue']
    for j, param_value in enumerate(parameter_values):
        # Update parameter value
        model_params[parameter_name] = param_value
        model = PINN().to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        # Train the model
        epochs = 1000
        for epoch in range(epochs):
            optimizer.zero_grad()
            loss = loss_function(model, eta_train, **model_params)
            loss.backward()
            optimizer.step()
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss: {loss.item()}')

        # Predict the solution
        eta_test = torch.tensor(np.linspace(0, 10, 100).reshape(-1, 1), dtype=torch.float32).to(device)
        y_pred = model(eta_test).detach().cpu().numpy()
        eta_test.requires_grad = True
        f_pred = model(eta_test)[:, 0]
        f_prime = torch.autograd.grad(f_pred, eta_test, grad_outputs=torch.ones_like(f_pred), create_graph=True)[0].detach().cpu().numpy()

        # Plot results
        plt.figure(1)
        plt.plot(eta_test.cpu().detach().numpy(), f_prime, label=f"f' ({parameter_name}={param_value})", color=colours[j])
        plt.xlabel('eta')
        plt.ylabel("f'")
        plt.legend()

        plt.figure(2)
        plt.plot(eta_test.cpu().detach().numpy(), y_pred[:, 1], label=f'g ({parameter_name}={param_value})', color=colours[j])
        plt.xlabel('eta')
        plt.ylabel('g')
        plt.legend()

        plt.figure(3)
        plt.plot(eta_test.cpu().detach().numpy(), y_pred[:, 2], label=f'h ({parameter_name}={param_value})', color=colours[j])
        plt.xlabel('eta')
        plt.ylabel('h')
        plt.legend()

    plt.show()

# Define parameter values for sweeps and run training/plotting for each parameter
parameter_sweeps = {
    'gamma': [0, 0.1, 0.5],
    'M': [0.5, 1, 2],
    'Pr': [0.71, 1, 2],
    'n': [0, 1, 2],
    'Nt': [0.1, 0.3, 0.5],
    'Nb': [0.1, 0.3, 0.5],
    'Ln': [0, 1, 2]
}

for parameter_name, values in parameter_sweeps.items():
    model_params = {'A': A, 'M': M, 'n': n, 'Pr': Pr, 'Nb': Nb, 'Nt': Nt, 'Ln': Ln, 'gamma': gamma, 'e_max': e_max}
    train_and_plot(values, parameter_name, model_params)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Check if GPU is available; use it if available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Define the neural network architecture
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.hidden_layer1 = nn.Linear(1, 128)
        self.hidden_layer2 = nn.Linear(128, 128)
        self.hidden_layer3 = nn.Linear(128, 128)
        self.output_layer = nn.Linear(128, 3)
        self.tanh = nn.Tanh()

    def forward(self, eta):
        x = self.tanh(self.hidden_layer1(eta))
        x = self.tanh(self.hidden_layer2(x))
        x = self.tanh(self.hidden_layer3(x))
        return self.output_layer(x)

# Define the ODE system
def ode_system(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max):
    eta.requires_grad = True
    y = model(eta)
    f, g, h = y[:, 0:1], y[:, 1:2], y[:, 2:3]

    # First, second, and third derivatives
    f_prime = torch.autograd.grad(f, eta, grad_outputs=torch.ones_like(f), create_graph=True)[0]
    f_double_prime = torch.autograd.grad(f_prime, eta, grad_outputs=torch.ones_like(f_prime), create_graph=True)[0]
    f_triple_prime = torch.autograd.grad(f_double_prime, eta, grad_outputs=torch.ones_like(f_double_prime), create_graph=True)[0]
    g_prime = torch.autograd.grad(g, eta, grad_outputs=torch.ones_like(g), create_graph=True)[0]
    g_double_prime = torch.autograd.grad(g_prime, eta, grad_outputs=torch.ones_like(g_prime), create_graph=True)[0]
    h_prime = torch.autograd.grad(h, eta, grad_outputs=torch.ones_like(h), create_graph=True)[0]
    h_double_prime = torch.autograd.grad(h_prime, eta, grad_outputs=torch.ones_like(h_prime), create_graph=True)[0]

    # ODEs
    ode1 = f_triple_prime + (n + 1) / 2 * f_prime * f_double_prime - n * f_prime ** 2 - M * f_prime - A * (f_prime + eta / 2 * f_double_prime)
    ode2 = g_double_prime + 0.5 * Pr * (n + 1) * f * g_prime + Nb * h_prime * g_prime + Nt * g_prime ** 2 - A / 2 * eta * g_prime
    ode3 = h_double_prime + (n + 1) / 2 * Ln * f * h_prime + Nt / Nb * g_double_prime - A * Ln / 2 * eta * h_prime
    return ode1, ode2, ode3

# Define the loss function
def loss_function(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max):
    eta.requires_grad = True
    y_pred = model(eta)
    f_pred, g_pred, h_pred = y_pred[:, :1], y_pred[:, 1:2], y_pred[:, 2:3]
    ode1, ode2, ode3 = ode_system(model, eta, A, M, n, Pr, Nb, Nt, Ln, gamma, e_max)

    # Boundary conditions at eta = 0 and eta = 10
    bc_eta_0 = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)
    bc_eta_10 = torch.tensor([[10.0]], dtype=torch.float32, requires_grad=True)
    bc_pred_0 = model(bc_eta_0)
    bc_pred_10 = model(bc_eta_10)

    f_bc0, g_bc0, h_bc0 = bc_pred_0[:, 0], bc_pred_0[:, 1], bc_pred_0[:, 2]
    f_prime_bc0 = torch.autograd.grad(f_bc0, bc_eta_0, grad_outputs=torch.ones_like(f_bc0), create_graph=True)[0]
    f_double_prime_bc0 = torch.autograd.grad(f_prime_bc0, bc_eta_0, grad_outputs=torch.ones_like(f_prime_bc0), create_graph=True)[0]
    f_prime_bc10 = torch.autograd.grad(bc_pred_10[:, 0], bc_eta_10, grad_outputs=torch.ones_like(bc_pred_10[:, 0]), create_graph=True)[0]

    # Boundary and ODE losses
    loss_bc = torch.mean((f_bc0 - 0) ** 2) + torch.mean((f_prime_bc0 - 1 - gamma * f_double_prime_bc0) ** 2) + \
              torch.mean((g_bc0 - 1) ** 2) + torch.mean((h_bc0 - 1) ** 2) + \
              torch.mean((f_prime_bc10 - 0) ** 2) + torch.mean((bc_pred_10[:, 1] - 0) ** 2) + \
              torch.mean((bc_pred_10[:, 2] - 0) ** 2)

    loss_ode = torch.mean(ode1 ** 2) + torch.mean(ode2 ** 2) + torch.mean(ode3 ** 2)

    return loss_ode + loss_bc

# Training data for eta
eta_train = torch.tensor(np.linspace(0, 10, 100).reshape(-1, 1), dtype=torch.float32, requires_grad=True)
A, M, n, Pr, Nb, Nt, Ln, gamma, e_max = 1, 0.5, 2, 1, 0.2, 0.2, 2, 0.1, 12

# Plotting function with loss tracking
def train_and_plot_loss(parameter_values, parameter_name, model_params):
    loss_history = {}  # Dictionary to store loss for each parameter value

    for param_value in parameter_values:
        model_params[parameter_name] = param_value
        model = PINN().to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        loss_epoch = []  # List to store loss at each epoch
        epochs = 1000
        for epoch in range(epochs):
            optimizer.zero_grad()
            loss = loss_function(model, eta_train, **model_params)
            loss.backward()
            optimizer.step()
            loss_epoch.append(loss.item())
            if epoch % 100 == 0:
                print(f'{parameter_name}={param_value} Epoch {epoch}, Loss: {loss.item()}')

        # Store the loss history for this parameter value
        loss_history[param_value] = loss_epoch

    # Plotting loss vs epochs for each parameter value
    plt.figure(figsize=(10, 6))
    for param_value, losses in loss_history.items():
        plt.plot(losses, label=f'{parameter_name}={param_value}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Loss vs Epochs for different values of {parameter_name}')
    plt.legend()
    plt.show()

# Define parameter values for sweeps and run training/plotting for each parameter
parameter_sweeps = {
    'gamma': [0, 0.1, 0.5],
    'M': [0.5, 1, 2],  # Replaced H with M
    'Pr': [0.71, 1, 2],
    'n': [0, 1, 2],
    'Nt': [0.1, 0.3, 0.5],
    'Nb': [0.1, 0.3, 0.5],
    'Ln': [0, 1, 2]
}

for parameter_name, values in parameter_sweeps.items():
    model_params = {'A': A, 'M': M, 'n': n, 'Pr': Pr, 'Nb': Nb, 'Nt': Nt, 'Ln': Ln, 'gamma': gamma, 'e_max': e_max}
    train_and_plot_loss(values, parameter_name, model_params)
```
